```python
from qualibrate import QualibrationNode
from quam_libs.iSWAP_t1t2 import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
node = QualibrationNode(name='iSWAP_t1t2_data')
```

```python
from pathlib import Path
data_path = Path('data/iSWAP_t1t2').resolve()
```

```python
idx = 7
t1_index, t2_index = 559, 560
t1_ds = node.load_from_id(t1_index, base_path=data_path).results['ds'].sel(qubit='q0_q2')
t2_ds = node.load_from_id(t2_index, base_path=data_path).results['ds'].sel(qubit='q0_q2')
```

```python
xnew, ynew = cos_conv_filter(t1_ds, idx, 5)
fig = exponetial(t1_ds, xnew, ynew, idx, 't1', True)
fig.savefig('paper_figure/iSWAP_t1_exponetial.pdf')
fig = rabi_chevron(t1_ds, xnew, ynew, idx, 't1')
fig.savefig('paper_figure/iSWAP_t1_rabi_chevron.pdf')
```

```python
xnew, ynew = cos_conv_filter(t2_ds, idx, 5)
fig = exponetial(t2_ds, xnew, ynew, idx, 't2', True)
fig.savefig('paper_figure/iSWAP_t2_exponetial.pdf')
fig = rabi_chevron(t2_ds, xnew, ynew, idx, 't2')
fig.savefig('paper_figure/iSWAP_t2_rabi_chevron.pdf')
```
